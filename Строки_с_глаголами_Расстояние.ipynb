```python
import pandas as pd
df = pd.read_csv('/content/odyss_verb_new.csv')
df
```

```python
row = df['row_num']

differences = [row[i] - row[i - 1] for i in range (1, len(row))]

average = sum(differences) / len(differences)
round(average, 2)
```
